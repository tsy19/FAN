```python
import matplotlib.pyplot as plt

# Load data from .txt file
with open('loss_prob_DP_sex.txt', 'r') as file:
    data = file.readlines()

# Remove any newline characters from the list
data = [float(line.strip()) for line in data]
```

```python
plt.plot(data)
plt.title("Training Loss: g(X) with Label $\omega_n$")
plt.xlabel("Rounds")
plt.ylabel("Loss")
plt.savefig("loss_g_x.png")
```
